### Sets

```python
# Sets are unordered lists that can be modified
set1 = {1,2,3}
print(set1)
```

```python
# Object type
type(set1)
```

```python
# We can use the set() function to create an object of this type. A set stores only unique values
set2 = set([1,2,3,2])
print(set2)
```

```python
type(set2)
```

```python
# We can store data of any type
set3 = {1.0,"Bob", (1,2,3)}
print(set3)
```

```python
# We can convert other objects to sets

# Empty set
print(set())

# String
print(set("Python"))

# Tuple
print(set(('a', 'e', 'i', 'o', 'u')))

# List
print(set(['a', 'e', 'i', 'o', 'u']))

# range
print(set(range(5)))
```

```python
# We can perform union or intersection aperations on sets
A = set(('a', 'e', 'i', 'o', 'u'))
B = set(['a', 2, 'z', 'o', (9,8)])

print("Union: ", A.union(B))
print("Intersection: ", A.intersection(B))
```

### List Comprehension
List comprehensions are highly optimized. Use them to filter data!

Calculate the squares of the numbers from 1 to 10

```python
# Without list comprehension

# Empty list to store the results 
squares_without_lc = []

# Loop through the range of numbers
for i in range(1,11):
    squares_without_lc.append(i * i)
print(squares_without_lc)
```

```python
# With list comprehensions
squares_with_lc = [i * i for i in range(1,11)]
print(squares_with_lc)
```

Print only the vowels in the sentence below

```python
# Phrase
Phrase = "The meteor was heading toward Earth"
```

```python
# List comprehensions
vowels = [i for i in Phrase if i in "aeiou"]
print(vowels)
```

Whrite down the square of each number from 1 to 10 and write down the number.

```python
# In this case, we use dictiornary comprehension
squares_dc = {i:i * i for i in range(1,11)}
print(squares_dc)
```

Convert the matrix below into a vector using list comprehension.

```python
# Matrix
matrix = [[0,0,0],[1,1,1], [2,2,2]]
print(matrix)
```

```python
#Vector
vector = [num for row in matrix for num in row]
print(vector)
```

```python
# Create a 3x4 matrix using nested list comprehensions
matrix = [[item for item in range(4)] for item in range(3)]
print(matrix)
```

 Match the word "code" with each item in the list below

```python
# List
cols = ['_A', '_B', '_C']
word = 'password'
```

### Regular Expressions

```python
import re
```

Python offer two different primitive operations based on regular expressions: 'match' checks for a match only at the beginning of the string, while 'search' checks for a match anywhere in the string. There are also several other pattern matching functions.

```python
# We check if the phrase begins with an alphanumeric character

# "^": This expression matches the start of a sequence
# "w+": This expression matches an alphanumeric character in the string

text1 = "AC56 is the most common code in the flight list."
reg1 = re.findall('^\w+', text1)
print(reg1)
```

```python
# We use\s to split a sentence by spaces 
print(re.split('\s', text1))
```

```python
# List the international flights
flights =  ['AC56 London', 'JK56 Paris', 'AC921 Philadelphia']

# We search for the fihgt codes for flight in flights:
for flight in flights:
    result = re.match('(^\w+)', flight)
    if result is not None:
        print(result.groups())
```

```python
# Pattern search

# List of patterns
words = ['Winter','London','Barcelona']

# Text
text = "AC56 is the most common code on the list of flights from London to Paris in winter"

# Search
for word in words:
    print('\nSearching for "%s" in the text ->' % (word) , end = ' ')
    if re.search(word, text):
        print('\nFound the pattern in the text!')
    else:
        print("Sorry, i didn't", word, 'in the text!')
```

```python
# Emails found

# Returns the emails found
list = 'bob@gmail.com, maria@hotmail.com, zicoyahoomail.com'
emails = re.findall('[\w\.-]+@[\w\.-]+', list)

# Returns the emails found
for email in emails:
    print('\nEmail found in the list: ', email)
```
